```python
import pandas as pd 
from tqdm import tqdm 
import transformers
from transformers import DistilBertTokenizerFast
from transformers import DistilBertForSequenceClassification
```

```python
df = pd.read_csv('cefr_leveled_texts.csv')
df.label = ['' for _ in range(len(df))]
df
```

```python
texts = df['text']
```

```python
from transformers import pipeline
pipe = pipeline('sentiment-analysis',model='results_2', tokenizer=DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased',config={'max_length':512, 'truncation':True, 'padding':True}))
```

```python
labels =  pipe(list(texts), max_length=512,truncation=True)
labels = [int(x['label'][-1]) for x in labels]
```

```python
df['label'] = labels 
df
```

```python
df.to_csv('cefr_bert_classified_2.csv',index=False)
```

```python
original = pd.read_csv('cefr_leveled_texts.csv')
original.label = original.label.apply(lambda x: x[0])
original.label = original.label.apply(lambda x: ['A','B','C'].index(x))
original
```

```python
from sklearn.metrics import accuracy_score,mean_absolute_error
accuracy_score(labels,list(original.label)),mean_absolute_error(labels,list(original.label))
```
