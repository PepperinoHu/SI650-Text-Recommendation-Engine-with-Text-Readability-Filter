```python
import pandas as pd 
import textstat
```

```python
df = pd.read_csv('onestop_split.csv') 
df
```

```python
df['Flesch Reading Ease'] = df['text'].apply(textstat.flesch_reading_ease)
df['Smog Index'] = df['text'].apply(textstat.smog_index)
df['Flesch Kincaid Grade'] = df['text'].apply(textstat.flesch_kincaid_grade) 
df['Coleman Liau Index'] = df['text'].apply(textstat.coleman_liau_index) 
df['Automated Readability Index'] = df['text'].apply(textstat.automated_readability_index)  
df['Dale Chall Readability Score'] = df['text'].apply(textstat.dale_chall_readability_score) 
df['Difficult Words'] = df['text'].apply(textstat.difficult_words)  
df['Gunning Fog'] = df['text'].apply(textstat.gunning_fog)  
df
```

```python
ele_stat = df[df.label==0].describe()
ele_stat
```

```python
int_stat = df[df.label==1].describe()
int_stat
```

```python
adv_stat = df[df.label==2].describe()
adv_stat
```

```python
comp = pd.DataFrame(columns = ele_stat.columns).append(ele_stat.loc['50%'])
comp = comp.append(int_stat.loc['50%'])
comp = comp.append(adv_stat.loc['50%'])
comp.set_index('label')
```

```python
import seaborn as sns
```

```python
comp
```

```python
comp['level'] = ['Elementary', 'Intermediate','Advanced']
```

```python
x = comp.copy()
x = x.T.reset_index().drop(9,axis = 0).drop(0,axis = 0)
x.columns = ['Indices','Ele','Int','Adv'] 
x = x.melt('Indices', var_name='Readability Indices',  value_name='Scores')
sns.set(rc = {'figure.figsize':(22,8)})
g = sns.barplot(x="Indices", y="Scores", hue='Readability Indices', data=x)
```

```python
x = comp.copy()
print(x)
x = x.drop('label',axis = 1)
x = x.melt('level', var_name='Readability Indices',  value_name='Scores')
sns.set(rc = {'figure.figsize':(15,8)})
g = sns.barplot(x="level", y="Scores", hue='Readability Indices', data=x)
```
