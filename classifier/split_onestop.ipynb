```python
import pandas as pd
```

```python
df = pd.read_csv('original_onestop.csv') 
df 
```

```python
new_df = pd.DataFrame(columns = df.columns)
for row in df.iterrows(): 
    l = row[1]['text'].split('\n')
    new_df.loc[len(new_df)] = [' '.join(l[:len(l)//2]),row[1]['label'] ]
    new_df.loc[len(new_df)] = [' '.join(l[len(l)//2:]),row[1]['label'] ]
new_df
```

```python
new_df.to_csv('onestop_split.csv',index=False)
```
